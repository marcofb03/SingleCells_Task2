## Loading Data

```python
import scanpy as sc
import numpy as np

#Loading KO files
seurat_int_KO = sc.read_h5ad("seurat_int_KO.h5ad")
seurat_int_KO.obsm['X_umap'] = np.array(seurat_int_KO.obsm['UMAP'])
seurat_int_KO.obsm['X_pca'] = np.array(seurat_int_KO.obsm['PCA'])

#Loading WT files
seurat_int_WT = sc.read_h5ad("seurat_int_WT.h5ad")
seurat_int_WT.obsm['X_umap'] = np.array(seurat_int_WT.obsm['UMAP'])
seurat_int_WT.obsm['X_pca'] = np.array(seurat_int_WT.obsm['PCA'])

# Umap whithout annotation
sc.pl.umap(seurat_int_WT, color=['seurat_clusters', 'sample'], wspace=0.4) 
sc.pl.umap(seurat_int_KO, color=['seurat_clusters', 'sample'], wspace=0.4)

# Umap with annotation
sc.pl.umap(seurat_int_WT, color='cell_type', wspace=0.4)
sc.pl.umap(seurat_int_KO, color='cell_type', wspace=0.4)

```
